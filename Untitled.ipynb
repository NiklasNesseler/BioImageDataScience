```python
def show_examples(image_list, title):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    for ax, img_path in zip(axes, image_list[:3]):
        img = imread(str(img_path))  # skimage reads tifs natively
        ax.imshow(img, cmap='gray')
        ax.set_title(img_path.name)
        ax.axis('off')
    fig.suptitle(title)
    plt.show()
```

```python
blurred = gaussian(img_src_norm, sigma=1.0)
```

```python
thresh = threshold_otsu(blurred)
binary = blurred > thresh
```

```python
from stardist.models import StarDist2D
from csbdeep.utils import normalize
from skimage.io import imread
from skimage.measure import regionprops_table
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = StarDist2D(None, name="cancer_model", basedir="Stardist_CancerCells_model")
labels_pred, _ = model.predict_instances(img_input)
```

```python
qa_source_dir = Path("Quality_control_data/230313_Stardist_CC_QA_round_source_JP")
qa_target_dir = Path("Quality_control_data/230313_Stardist_CC_QA_round_target_JP")
qa_results = []
all_features = []
```

```python
for src_path in sorted(qa_source_dir.glob("*.tif")):
    img = imread(src_path)
    img_input = normalize(img, 1, 99.8)
    labels_pred, _ = model.predict_instances(img_input)

    # Save prediction
    pred_name = src_path.stem + "_pred.tif"
    imsave(f"output/{pred_name}", labels_pred.astype(np.uint16))

    # Load ground truth
    gt_path = qa_target_dir / src_path.name
    if gt_path.exists():
        gt = imread(gt_path)
        ari, _, _ = adapted_rand_error(gt, labels_pred)
        voi_split, voi_merge = variation_of_information(gt, labels_pred)
        qa_results.append({
            "filename": src_path.name,
            "adapted_rand_index": 1 - ari,
            "voi_split": voi_split,
            "voi_merge": voi_merge
        })

    # Compute features
    props = regionprops_table(labels_pred, intensity_image=img_input,
        properties=["label", "area", "eccentricity", "solidity", "mean_intensity"])
    df_props = pd.DataFrame(props)
    df_props["image"] = src_path.name
    all_features.append(df_props)
```

```python
qa_df = pd.DataFrame(qa_results)
qa_df.to_csv("output/qa_evaluation.csv", sep=';', index=False)
```

```python
features_df = pd.concat(all_features, ignore_index=True)
features_df.to_csv("output/qa_features_summary.csv", sep=';', index=False)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
qa_df.plot(kind='bar', x='filename', y='adapted_rand_index', ax=axes[0], color='skyblue')
axes[0].set_title("Adapted Rand Index")
qa_df.plot(kind='bar', x='filename', y='voi_split', ax=axes[1], color='salmon')
axes[1].set_title("VOI Split")
qa_df.plot(kind='bar', x='filename', y='voi_merge', ax=axes[2], color='seagreen')
axes[2].set_title("VOI Merge")
plt.tight_layout()
plt.show()
```

```python
agg = features_df.groupby("image").agg({"area": "mean", "mean_intensity": "mean"})
agg.plot(kind="bar", figsize=(10, 5))
plt.title("Mean Area and Intensity per QA Image")
plt.ylabel("Value")
plt.tight_layout()
plt.show()
```

```python
def iou_score(seg, gt):
    seg_bin = seg > 0   # alles was nicht Hintergrund ist
    gt_bin = gt > 0
    intersection = np.logical_and(seg_bin, gt_bin).sum()
    union = np.logical_or(seg_bin, gt_bin).sum()
    return intersection / union if union > 0 else 0

iou = iou_score(labeled, gt)
print(f"IoU der Segmentierung vs. Ground Truth: {iou:.3f}")
```
