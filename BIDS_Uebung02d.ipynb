```python
import numpy as np
import stackview
from skimage.data import cells3d
```

```python
image = cells3d()[30,0]
image.shape
```

```python
stackview.imshow(image)
```

```python
min_value = image.min()
max_value = image.max()
min_value, max_value
```

```python
stackview.imshow(image, max_display_intensity=10000)
```

```python
nuclei = cells3d()[30,1]
nuclei.shape
```

```python
stackview.imshow(nuclei)
```

```python
upper_limit = np.percentile(image, 99)
upper_limit
stackview.imshow(nuclei, max_display_intensity=upper_limit)
```
